```
datos = Import["cnae9_10NN.mtx", "MTX"]
```

```
(* Procesar los datos (excluir la primera fila) *)
aristas = Rest[datos];

(* Crear el grafo con aristas ponderadas *)
grafo = Graph[
  aristas[[All, {1, 2}]], (* Conexiones entre nodos *)
  EdgeWeight -> aristas[[All, 3]] (* Pesos en las aristas *)
];
```
